```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get('https://archive.ics.uci.edu/')
```

```python
viewDatasets= driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()
```

```python
datasetAnchor = driver.find_elements(By.XPATH,'//div[@class="flex flex-col gap-1"]/div[@role="row"]/div/div[2]/h2/a')
url_list=[]
for urlElement in datasetAnchor:
    url_list.append(urlElement.get_attribute('href'))
```

```python
print(len(url_list))
```

```python
Dname=[]
Dtype=[]
Dtask=[]
Dfeature=[]
Dnooffeature=[]
Dnoofinstance=[]
Dyear=[]
```

```python
for url in url_list:
    driver.get(url)
    DatasetnameEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1').text
    DtaskEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p').text
    DfeatureEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p').text
    DnoOfFeatureEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p').text
    DnoOfinstanceEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p').text
    wait = WebDriverWait(driver,5)
    try :
        DatasetYearEle = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2'))).text
    except:
        DatasetYearEle = 'abcdefghi-'
    #     print(DatasetYearEle)
        #     DatasetYearEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
    DatasetTypeEle = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p').text
    
    Dname.append(DatasetnameEle)
    Dtype.append(DatasetTypeEle)
    Dtask.append(DtaskEle)
    Dfeature.append(DfeatureEle)
    Dnooffeature.append(DnoOfFeatureEle)
    Dnoofinstance.append(DnoOfinstanceEle)
    Dyear.append(DatasetYearEle[11:])
```

```python
print(Dyear)
```

```python
data = {
    'dataset name':Dname,
    'dataset year':Dyear,
    'dataset type':Dtype,
    'task':Dtask,
    'attribute type':Dfeature,
    'no. of attribute':Dnooffeature,
    'no. of instances':Dnoofinstance
}

dataSetDataFrame = pd.DataFrame(data=data)
dataSetDataFrame
```
