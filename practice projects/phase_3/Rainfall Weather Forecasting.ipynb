# Rainfall Weather Forecasting

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error ,r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df_rainfall=pd.read_csv('Rainfall.csv')
```

```python
df_rainfall
```

```python
df_rainfall.isnull().sum()
```

```python
columns = df_rainfall.columns[2:]
```

```python
columns
```

```python
for col in columns:
    print(col)
    if col != 'WindGustDir' and col != 'WindDir9am' and col != 'WindDir3pm' and col != 'RainToday' and col != 'RainTomorrow':
        median = df_rainfall[col].median()
        df_rainfall[col].fillna(median, inplace=True)
```

```python
df_rainfall
```

```python
df_rainfall.isnull().sum()
```

```python
df_rainfall.isnull().sum()
```

```python
df_rainfall.dtypes
```

```python
for col in df_rainfall.select_dtypes(include=['object']).columns:
    mode = df_rainfall[col].mode()[0]
    df_rainfall[col].fillna(mode, inplace=True)
```

```python
df_rainfall
```

```python
df_rainfall.isnull().sum()
```

```python
label_encoder = LabelEncoder()
```

```python
df_rainfall['WindGustDir'] = label_encoder.fit_transform(df_rainfall['WindGustDir'])
```

```python
df_rainfall['WindDir9am'] = label_encoder.fit_transform(df_rainfall['WindDir9am'])
```

```python
df_rainfall['WindDir3pm'] = label_encoder.fit_transform(df_rainfall['WindDir3pm'])
```

```python
df_rainfall['RainToday'] = label_encoder.fit_transform(df_rainfall['RainToday'])
```

```python
df_rainfall['RainTomorrow'] = label_encoder.fit_transform(df_rainfall['RainTomorrow'])
```

```python
df_rainfall['Location'] = label_encoder.fit_transform(df_rainfall['Location'])
```

```python
df_rainfall
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df_rainfall.iloc[:,1:-2]
```

```python
x
```

```python
y=df_rainfall.iloc[:,-1]
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred_y = model.predict(x_test)
```

```python
pred_y
```

```python
r2Score = r2_score(y_test,pred_y)
```

```python
print("R2 Score :: ",round(r2Score,2))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model_max = LinearRegression()
```

```python
param_grids_max = {
    'fit_intercept': [True, False]
}
```

```python
grid_search_max = GridSearchCV(model_max,param_grids_max , cv=5, scoring='neg_mean_squared_error')
```

```python
grid_search_max.fit(x_train, y_train)
```

```python
best_model_max = grid_search_max.best_estimator_
best_params_max = grid_search_max.best_params_
```

```python
print("Best hyperparameters:", best_params_max)
```

```python
best_model_max.fit(x_train, y_train)
```

```python
pred_y_max = best_model_max.predict(x_test)
```

```python
best_score_max = grid_search_max.best_score_
print("Best score:", best_score_max)
```

```python
positive_mse_max = -best_score_max
```

```python
print("Positive mean squared error:", positive_mse_max)
```
