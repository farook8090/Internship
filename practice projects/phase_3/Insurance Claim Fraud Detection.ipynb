```python
column_names = [
    'months_as_customer',
    'age',
    'policy_number',
    'policy_bind_date',
    'policy_state',
    'policy_csi',
    'policy_deductable',
    'policy_annual_premium',
    'umbrella_limit',
    'insured_zip',
    'insured_sex',
    'insured_education_level',
    'insured_occupation',
    'insured_hobbies',
    'insured_relationship',
    'capital-gains',
    'capital-loss',
    'incident_date',
    'incident_type',
    'collision_type',
    'incident_severity',
    'authorities_contracted',
    'incident_state',
    'incident_city',
    'incident_location',
    'incident_hour_of_the_day',
    'number_of_vehicles_involved',
    'property_damage',
    'bodily_injuries',
    'witenesses',
    'police_report_available',
    'total_claim_amount',
    'injury_claim',
    'property_claim',
    'vehicle_claim',
    'auto_make',
    'auto_model',
    'auto_year',
    'fraud_reported'
]
```

```python
print(len(column_names))
```

```python
df_auto.columns = column_names
```

```python
df_auto
```

```python
label_encoder = LabelEncoder()
```

```python
for column in df_auto.columns:
    if df_auto[column].dtype == 'object':
        df_auto[column] = label_encoder.fit_transform(df_auto[column])
```

```python
df_auto
```

```python
x = df_auto.iloc[:,0:-1] #feature variable
```

```python
x
```

```python
y = df_auto.iloc[:,-1] #target variable
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
```

```python
LR = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
LR.fit(X_train,y_train)
```

```python
pred_y = LR.predict(X_test)
```

```python
MSE = mean_squared_error(y_test,pred_y)
```

```python
print("MSE :: ",round(MSE,2))
```

Hyperparameter tuning

```python
model_max = LinearRegression()
```

```python
param_grids_max = {
    'fit_intercept': [True, False]
}
```

```python
grid_search_max = GridSearchCV(model_max,param_grids_max , cv=5, scoring='neg_mean_squared_error')
```

```python
grid_search_max.fit(X_train, y_train)
```

```python
best_model_max = grid_search_max.best_estimator_
best_params_max = grid_search_max.best_params_
```

```python
print("Best hyperparameters:", best_params_max)
```

```python
best_model_max.fit(X_train, y_train)
```

```python
pred_y_max = best_model_max.predict(X_test)
```

```python
best_score_max = grid_search_max.best_score_
print("Best score:", best_score_max)
```

```python
positive_mse_max = -best_score_max
```
