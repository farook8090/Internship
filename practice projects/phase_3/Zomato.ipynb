```python
df=pd.read_csv('zomato.csv')
```

```python
df.isnull().sum()
```

```python
df.fillna({
    'Restaurant ID': df['Restaurant ID'].median(),
    'Restaurant Name': df['Restaurant Name'].mode()[0],
    'City': df['City'].mode()[0],
    'Address': df['Address'].mode()[0],
    'Locality': df['Locality'].mode()[0],
    'Longitude': df['Longitude'].median(),
    'Latitude': df['Latitude'].median(),
    'Cuisines': df['Cuisines'].mode()[0],
    'Average Cost for two': df['Average Cost for two'].median(),
    'Price range': df['Price range'].median(),
    'Aggregate rating': df['Aggregate rating'].median(),
    'Votes': df['Votes'].median(),
    'Locality Verbose': df['Locality Verbose'].mode()[0],
    'Currency': df['Currency'].mode()[0],
    'Has Table booking': df['Has Table booking'].mode()[0],
    'Has Online delivery': df['Has Online delivery'].mode()[0],
    'Is delivering now': df['Is delivering now'].mode()[0],
    'Switch to order menu': df['Switch to order menu'].mode()[0],
    'Rating color': df['Rating color'].mode()[0],
    'Rating text': df['Rating text'].mode()[0],
    'Country': df['Country'].mode()[0],
}, inplace=True)
print('Missing values handled.')
```

```python
df.isnull().sum()
```

```python
categorical_cols = ['City', 'Cuisines', 'Rating text']
numerical_cols = ['Longitude', 'Latitude', 'Average Cost for two', 'Aggregate rating', 'Votes']

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ]
)

df_transformed = preprocessor.fit_transform(df)
print('Data encoded and scaled.')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
X_train, X_test, y_train_cost, y_test_cost = train_test_split(df_transformed, df['Average Cost for two'], test_size=0.2, random_state=42)
X_train, X_test, y_train_range, y_test_range = train_test_split(df_transformed, df['Price range'], test_size=0.2, random_state=42)

regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train, y_train_cost)

classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, y_train_range)

print('Models trained.')
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
y_pred_cost = regressor.predict(X_test)
rmse = mean_squared_error(y_test_cost, y_pred_cost, squared=False)

y_pred_range = classifier.predict(X_test)
accuracy = accuracy_score(y_test_range, y_pred_range)

print(f'RMSE for Average Cost for two: {rmse}')
print(f'Accuracy for Price range: {accuracy}')
```
