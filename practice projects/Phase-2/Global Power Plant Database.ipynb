Model Preparation start

Primary Fuel Model Preparation start

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
X_fuel = df.drop(columns=['primary_fuel'])  # Features Variabes
y_fuel = df['primary_fuel']  # Target variable
```

```python
label_encoder = LabelEncoder()
y_fuel_encoded = label_encoder.fit_transform(y_fuel)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_fuel, y_fuel_encoded, test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)  # Initialize Random Forest classifier
rf_classifier.fit(X_train, y_train)  # Train the classifier
```

```python
y_pred_fuel_encoded = rf_classifier.predict(X_test)
```

```python
y_pred_fuel = label_encoder.inverse_transform(y_pred_fuel_encoded)
```

```python
print("Predicted fuel names:", y_pred_fuel)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

Primary Fuel Model Preparation end

```python
label_encoder = LabelEncoder()
```

```python
df['primary_fuel'] = label_encoder.fit_transform(df['primary_fuel'])
```

```python
X_capacity = df.drop(columns=['capacity_mw'])  # Features Variabes
y_capacity = df['capacity_mw']  # Target variable
```

```python
X_train_capacity, X_test_capacity, y_train_capacity, y_test_capacity = train_test_split(X_capacity, y_capacity, test_size=0.2, random_state=42)
```

```python
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)  # Initialize Random Forest regressor
rf_regressor.fit(X_train_capacity, y_train_capacity)  # Train the regressor
```

```python
y_pred_capacity = rf_regressor.predict(X_test_capacity)
```

```python
mse = mean_squared_error(y_test_capacity, y_pred_capacity)
print("Mean Squared Error:", mse)
```
