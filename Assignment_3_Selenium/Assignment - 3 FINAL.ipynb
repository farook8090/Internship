```python
driver = webdriver.Chrome()
driver.get('https://www.amazon.in/')

delay = 10
# WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")))
search_bar = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search_bar.send_keys("guitar")

search_icon = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search_icon.click()

product_urls_list = []
name = []
price = []
brand = []
return_policy = []
delivery_date = []
availability = []
url = []
start = 0
end = 3

for page in range(start,end):
    url_element = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for url in url_element:
        product_urls_list.append(url.get_attribute('href'))    
#     nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    
    nxt_button.click()
    time.sleep(2)

print(len(product_urls_list))
```

```python
print(len(product_urls_list))
```

```python
for url in product_urls_list:
        driver.get(url)
        time.sleep(2)

        try:
            product_name = driver.find_element(By.XPATH,'//*[@id="title_feature_div"]/div/h1/span').text
            product_price = driver.find_element(By.XPATH,'//*[@id="corePrice_feature_div"]/div/div/span[1]/span[2]/span[2]').text
            product_brand = driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span').text
            product_return = driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span').text
            product_delivery = driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]').text
            product_availability = driver.find_element(By.XPATH,'//*[@id="availability"]/span').text

            name.append(product_name)
            price.append(product_price)
            brand.append(product_brand)
            return_policy.append(product_return)
            delivery_date.append(product_delivery)
            availability.append(product_availability)
        except NoSuchElementException as e:
#             print("Error :: ",e)
            name.append("-")
            price.append("-")
            brand.append("-")
            return_policy.append("-")
            delivery_date.append("-")
            availability.append("-")
```

```python
print(brand)
```

```python
#Making all arrays of same length

max_lenth = max(len(product_urls_list),len(name),len(price),len(brand),len(return_policy),len(delivery_date),len(availability))
# max_lenth

product_urls_list += ["-"] * (max_lenth - len(product_urls_list))
name += ["-"] * (max_lenth - len(name))
price += ["-"] * (max_lenth - len(price))
brand += ["-"] * (max_lenth - len(brand))
return_policy += ["-"] * (max_lenth - len(return_policy))
delivery_date += ["-"] * (max_lenth - len(delivery_date))
availability += ["-"] * (max_lenth - len(availability))
```

```python
data={
    "url":product_urls_list,
    "name":name,
    "price":price,
    "brand":brand,
    "return_policy":return_policy,
    "delivery_date":delivery_date,
    "availability":availability
}
product_df = pd.DataFrame(data=data)
product_df.head(15)
```
