```python
state_2001 = crimes_2001_pop_lit['State']
population_2001 = crimes_2001_pop_lit['Population']
crimesT_2001 = crimes_2001_pop_lit['Total Crimes']
```

```python
df_2001_5 = {
    "state":stateList,
    "population":populationlist,
    "crime":crimeTotalList
}
```
